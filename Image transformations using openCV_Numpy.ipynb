# Converting an RGB image to HSV

```python
 def f_rgb_to_hsv(r, g, b, scaleFactor):
	r, g, b = r / 255.0, g / 255.0 , b / 255.0 # normalizing r, g, b values
	cmax = max(r, g, b) # max of r, g, b 
	cmin = min(r, g, b) # min of r, g, b
	diff = cmax - cmin  # diff of cmax and cmin.
	
	if cmax == cmin:
		h = 0
	elif cmax == r:
		h = (60 * ((g - b) / diff) + 0) % 360
	elif cmax == g:
		h = (60 * ((b - r) / diff) + 120) % 360
	elif  cmax == b:
		h = (60 * ((r - g) / diff) + 240) % 360
		
	if h < 0:
		h = h + 360
	if cmax == 0:
		 s = 0
	else:
		s = (diff / cmax) * scaleFactor
	v = cmax * scaleFactor
	
	return h, s, v 

f_rgb_to_hsv(100, 200, 50, 100)

#scaleFactor: If scaleFactor to be 100 then that means saturation(s) and value(v) will be between 0 to 100 
```

# OpenCV method to convert a BGR to HSV image

```python
im = cv2.imread(r'D:\Study_materials\Machine_Learning\images\bird.jpg')
hsvim = cv2.cvtColor(im, cv2.COLOR_BGR2HSV) # converting BGR image to HSV

bgrim = cv2.cvtColor(hsvim, cv2.COLOR_HSV2BGR) # converting hsv image to bgr
rgbim = cv2.cvtColor(hsvim, cv2.COLOR_HSV2RGB) # converting hsv image to rgb

# displaying hsv image
cv2.imshow('HSV Color image',hsvim)
cv2.waitKey(0)
cv2.destroyAllWindows()

# displaying rgb image
cv2.imshow('RGB Color image',rgbim)
cv2.waitKey(0)
cv2.destroyAllWindows()

# displaying bgr image
cv2.imshow('BGR Color image',bgrim)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# segmentation of red color

```python
import numpy as np
img = cv2.imread(r'D:\Study_materials\Machine_Learning\images\flowers.jpg')
# converting BGR image to HSV 
imghsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lr = np.array([0, 120, 70])
ur= np.array([15, 255, 255])
mask1 = cv2.inRange(imghsv, lr, ur)

lr = np.array([165, 120, 70])
ur= np.array([180, 255, 255])
mask2 = cv2.inRange(imghsv, lr, ur)

mask = mask1 | mask2

# performing bitwise-and operation on two source images and using mask value
resultant_image = cv2.bitwise_and(imghsv, imghsv, mask = mask)
# converting resultant HSV image to RGB
res_image_RGB = cv2. cvtColor(resultant_image, cv2.COLOR_HSV2RGB) 
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:, :, ::-1]) # original image converting RGB to BGR 

plt.subplot(122)
plt.imshow(res_image_RGB) 
plt.show()
```
